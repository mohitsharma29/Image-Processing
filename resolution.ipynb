```python
grayImg = np.zeros((256,256))
for i in range(256):
    for j in range(256):
        grayImg[i][j] = round((int(img[i][j][0]) + int(img[i][j][1]) + int(img[i][j][2]))/3)
```

```python
plt.imshow(grayImg)
```

```python
# Spatial Resolution
def spatial(img):
    globalMat = img
    while img.shape != (1,1):
        m= 0
        n = 0
        tempMat = np.zeros((int(globalMat.shape[0]/2), int(globalMat.shape[1]/2)))
        for i in range(0,globalMat.shape[0],2):
            n = 0
            for j in range(0,globalMat.shape[0],2):
                tempMat[m][n] = globalMat[i][j]
                n += 1
            m += 1
        globalMat = tempMat
        plt.figure()
        plt.imshow(globalMat)
        if globalMat.shape == (1,1):
            break
```

```python
spatial(grayImg)
```
